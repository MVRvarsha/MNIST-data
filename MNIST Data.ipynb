```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import precision_score, recall_score, f1_score

# Load and preprocess MNIST dataset
def load_and_preprocess_data():
    (x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize pixel value to range[0,1]
    x_train = x_train.astype('float32') / 255.0
    x_test = x_test.astype('float32') / 255.0

#Reshape data for FCNNs
    x_train_flat = x_train.reshape(-1, 28*28)
    x_test_flat = x_test.reshape(-1, 28*28)

#Reshape data for CNNs
    x_train_cnn = x_train.reshape(-1, 28, 28, 1)
    x_test_cnn = x_test.reshape(-1, 28, 28, 1)

#Convert lables to one-hot encoding
    y_train = tf.keras.utils.to_categorical(y_train, 10)
    y_test = tf.keras.utils.to_categorical(y_test, 10)
    return (x_train_flat, x_train_cnn, y_train), (x_test_flat, x_test_cnn, y_test)

#Traning the models
#1. Shallow Fully Connected Neural Network (FCNN):
def create_shallow_fcnn():
    model = models.Sequential([
        layers.Input(shape=(784,)),
        layers.Dense(128, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])
    return model

#2. Deep Fully Connected Neural Network (FCNN):
def create_deep_fcnn():
    model = models.Sequential([
        layers.Input(shape=(784,)),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(10, activation='softmax')
    ])
    return model

#3. Shallow Convolutional Neural Network (CNN):
def create_shallow_cnn():
    model = models.Sequential([
        layers.Input(shape=(28, 28, 1)),
        layers.Conv2D(8, (3, 3), padding='same', activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])
    return model

#4. Deep Convolutional Neural Network (CNN):
def create_deep_cnn():
    model = models.Sequential([
        layers.Input(shape=(28, 28, 1)),
        layers.Conv2D(16, (3, 3), padding='same', activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(32, (3, 3), padding='same', activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(10, activation='softmax')
    ])
    return model

#training and evaluating the model
def train_and_evaluate_model(model, x_train, y_train, x_test, y_test, model_name):
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
    print(f"\nModel: {model_name}")
    model.summary()
    start_time = time.time()
    history = model.fit(x_train, y_train, batch_size=64, epochs=20, validation_data=(x_test, y_test), verbose=1)
    training_time = time.time() - start_time
    test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
    return history, test_accuracy, training_time, model

#Evaluation of the model
def evaluate_model(model, x_test, y_test, model_name):
    y_pred = model.predict(x_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true = np.argmax(y_test, axis=1)

    # Calculate metrics
    precision = precision_score(y_true, y_pred_classes, average='macro')
    recall = recall_score(y_true, y_pred_classes, average='macro')
    f1 = f1_score(y_true, y_pred_classes, average='macro')

    # Print the results
    print(f"\nEvaluation Results for {model_name}:")
    print("-" * 80)
    print(f"{'Metric':<15} {'Value':<15}")
    print("-" * 80)
    print(f"{'Precision':<15} {precision:>6.2f}")
    print(f"{'Recall':<15} {recall:>6.2f}")
    print(f"{'F1-Score':<15} {f1:>6.2f}")
    print("-" * 80)

#plotting the confusion matrix
def plot_confusion_matrix(model, x_test, y_test, model_name):
    y_pred = model.predict(x_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true = np.argmax(y_test, axis=1)
    cm = confusion_matrix(y_true, y_pred_classes)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.show()

def plot_history(histories, titles):
    for history, title in zip(histories, titles):
        plt.figure(figsize=(12, 4))
        plt.subplot(1, 2, 1)

        plt.plot(history.history['accuracy'], label='Training Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.title(f'{title} - Accuracy')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title(f'{title} - Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.show()

#loading and preprocessing the models
def main():
    (x_train_flat, x_train_cnn, y_train), (x_test_flat, x_test_cnn, y_test) = load_and_preprocess_data()
    models_config = [
        (create_shallow_fcnn(), x_train_flat, x_test_flat, "Shallow FCNN"),
        (create_deep_fcnn(), x_train_flat, x_test_flat, "Deep FCNN"),
        (create_shallow_cnn(), x_train_cnn, x_test_cnn, "Shallow CNN"),
        (create_deep_cnn(), x_train_cnn, x_test_cnn, "Deep CNN")
    ]
    histories = []
    results = []
    for model, x_train, x_test, name in models_config:
        print(f"\nTraining {name}...")
        history, test_acc, train_time, trained_model = train_and_evaluate_model(model, x_train, y_train, x_test, y_test, name)
        histories.append(history)
        results.append({'name': name, 'test_accuracy': test_acc, 'training_time': train_time, 'parameters': model.count_params()})
        plot_confusion_matrix(trained_model, x_test, y_test, name)

        # Evaluate the model
        evaluate_model(trained_model, x_test, y_test, name)

    # Printing the final results
    print("\nFinal Results:")
    print("-" * 80)
    print(f"{'Model':<15} {'Test Accuracy':<15} {'Training Time':<15} {'Parameters':<15}")
    print("-" * 80)
    for result in results:
        print(f"{result['name']:<15} {result['test_accuracy']*100:>6.2f}%{' '*8} {result['training_time']:>6.1f}s{' '*8} {result['parameters']:>10,d}")

    # Plotting the training history
    plot_history(histories, [model['name'] for model in results])

if __name__ == "__main__":
    main()
```
